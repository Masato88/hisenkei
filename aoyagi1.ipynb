```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#t=0~20でdt=1
dt = 1
T =10
Nt= int(T/dt)+1
t = np.linspace(0,T,Nt)
sigma = np.zeros((3,7,Nt))
m = np.zeros((3,7,Nt))
a = np.array([0.02,0.05,0.1])
U = np.zeros((3,7,Nt))
q = np.ones(Nt)
#初期値
for k in range(3):
  for i in range(7):
    m[k][i][0] = 0.5+0.05*i
    sigma[k][i][0] = np.sqrt(a[k])
```

```python
#漸化式
for j in range(3): #aのループ
  for k in range(7): # mのループ
    for i in range(Nt-1): #tのループ
      U[j][k][i+1] = np.sqrt(2/(math.pi*sigma[j][k][i]**2))*math.exp(-0.5*m[j][k][i]**2/sigma[j][k][i]**2)
      m[j][k][i+1] = math.erf(m[j][k][i]/(sigma[j][k][i]*math.sqrt(2)))
      sigma[j][k][i+1] = math.sqrt(a[j]*q[i+1] + (U[j][k][i+1]**2)*(sigma[j][k][i]**2) + 2*a[j]*U[j][k][i+1]*m[j][k][i]*m[j][k][i+1])
      #print()
```

```python
fig = plt.figure
for i in range(7):
  plt.plot(t,m[0][i])

plt.title('overlap a=0.02') # グラフのタイトル
plt.xticks(t)
plt.xlabel('t') #x軸のラベル
plt.ylabel('overlap') #y軸のラベル
plt.legend()
```
